```python
from scipy.special import erfc
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
```

```python
def concentration_1d(t,x,c0,c1,D):
    return c0+(c1-c0)*erfc(x/(2*sqrt(D*t)))
```

```python
xrange = np.arange(0, 5, .1) #0 to 5 nm
trange = np.linspace(0,6500,10)
c0 = 0.23   #Concentration of infinite reservoir of Cr
c1 = 0  # Interface concentration of Cr
Dcr = 1*10**-4 #Interdiffusion coefficeint of Cr in nm^2/s

plt.plot(xrange, concentration_1d(1,xrange,c0,c1,Dcr), label='t=1')
plt.plot(xrange, concentration_1d(10,xrange,c0,c1,Dcr), label='t=10')
plt.plot(xrange, concentration_1d(1000,xrange,c0,c1,Dcr), label='t=1000')
plt.plot(xrange, concentration_1d(6500,xrange,c0,c1,Dcr), label='t=6500')
plt.xlabel("distance (nm)")
plt.ylabel("Cr fraction (wt%)")
plt.legend()

print(np.average(concentration_1d(6500,xrange,c0,c1,Dcr)[:25]))
```

```python
drange = [1*10**-5,1*10**-4,1*10**-3] #nm^2/s
drange_cm2 = [1e-15,1e-14,1e-13] #cm^2/s
for i,d in enumerate(drange):
    plt.plot(xrange, concentration_1d(6500,xrange,c0,c1,d),label=drange_cm2[i])

plt.legend()
plt.ylabel("Cr concentration")
plt.xlabel("Distance from oxide interface (nm)")
```
