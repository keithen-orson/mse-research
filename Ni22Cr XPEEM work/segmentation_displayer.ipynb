```python
#Read an image and a label, along with a prediction
image = Image.open('image_151.tif')
labels = Image.open('image_151 2.tif')
prediction = Image.open("/Users/apple/MLIAFinalProject/predictions/TU_Synapse64/TU_pretrain_R50-ViT-B_16_skip3_epo900_bs24_64_s69420/augmented_85_pred.tif")


image = np.array(image)
label_image = np.array(labels,dtype=int)
pred_image = np.array(prediction, dtype=int)
#print(label_image.shape)
print(distance.dice(np.ndarray.flatten(label_image),np.ndarray.flatten(pred_image)))

# to make the background transparent, pass the value of `bg_label`,
# and leave `bg_color` as `None` and `kind` as `overlay`
image_label_overlay = label2rgb(label_image, image=image, bg_label=0)
image_prediction_overlay = label2rgb(pred_image,image=image, bg_label=0)

#Print the regionprops from the label function in skimage 
skimagelabeled = label(label_image)
regions = regionprops(skimagelabeled)

```
