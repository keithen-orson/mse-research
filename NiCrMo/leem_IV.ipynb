```python
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib as mpl
import numpy as np
mpl.rcParams['figure.dpi'] = 100
```

```python
iv_curves = pd.read_csv('/Users/apple/Sync/Research/Mo XPEEM/LEEM IV curves.csv')
iv_curves.columns
```

```python

plt.plot(iv_curves['Start Energy'],iv_curves['Left Grain'])
plt.plot(iv_curves['Start Energy'],iv_curves['Center Grain'])
plt.plot(iv_curves['Start Energy'],iv_curves['Right Grain'])
plt.ylabel("Intensity (counts/s)")
plt.xlabel("Start energy (eV)")
plt.legend(["Left Grain","Center Grain","Right Grain"])
```

```python
plt.plot(iv_curves['Start Energy'],np.gradient(iv_curves['Left Grain']))
plt.plot(iv_curves['Start Energy'],np.gradient(iv_curves['Center Grain']))
plt.plot(iv_curves['Start Energy'],np.gradient(iv_curves['Right Grain']))
plt.ylabel("Derivative of Intensity DI/DE (counts/eVs)")
plt.xlabel("Start energy (eV)")
plt.legend(["Left Grain","Center Grain","Right Grain"])
```

```python
for column,series in iv_curves.iteritems():
    print(column)
    print(np.argmin(np.gradient(series)))
```
